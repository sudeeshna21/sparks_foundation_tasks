```python
#importing modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading data
data=pd.read_csv('http://bit.ly/w-data')
data.head()
```

```python
data.shape
```

```python
# visualize data
data.plot(x='Hours',y='Scores',style='o')
plt.xlabel('Hours studied')
plt.ylabel('score of students')
plt.show()
```

As the no of hours increases score is also increasing

```python
# separating dependent and independent variables
x=data.iloc[:, :-1].values  
y=data.iloc[:, 1].values  
```

```python
# splitting data into train and test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.25, random_state=0) 
```

```python
# training the data
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
#predicting the values
y_pred=lin_reg.predict(x_test)
```

```python
import seaborn as sns
sns.jointplot(x=y_test,y=y_pred,kind="regg", color="Blue")
```

we can see that y_pred and y_test are correlated that means the model is working good.

```python
# training score and testing score
sc=lin_reg.score(x_train,y_train)
print('training score:',sc)
test_sc=lin_reg.score(x_test,y_test)
print('test score:',test_sc)
```

```python
# mean absolute error
from sklearn import metrics  
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred)) 
```

```python
# prediction for 9.25 hrs
hours = 9.25
pred = lin_reg.predict([[hours]])
print("No of Hours = ",hours)
print("Predicted Score = ",pred[0])
```

# conclusion:
After training the model I have predicted score of student who studies for 9.25hr/day and the score is 93.8927.On evaluating the model I got Mean absolute error of 4.13.
